#### 1. Count words in a text

a. Output a list of words in the file along with their frequency counts (ignoring case).   
b. Count how many unique words there are (ignoring case).    
c. Check how common are all different sequences of vowels (e.g. the sequences "ieu" or just "e" in "lieutenant")?

```python
# a)
import string

#a function to separate out non-letter characters
def letters(word1):
    not_let = 0
    for letter1 in word1:
        if letter1 not in string.ascii_letters:
            return False
    return True

# Separate words and filter out punctuation and digits

words = []
[words.append(word.strip(string.punctuation).lower()) for word in text.split() if letters(word)==True]



#Make a function that returns a dictionary of words and their frequency counts
#def counter(w):

def counter(i, count_dict):
    if i in count_dict:
        count_dict[i] += 1
    else:
        count_dict[i] = 1
    return count_dict

count = {}
[counter(word,count) for word in words]
print(count)
```

```python
# b)
unique_words = len(count)

print(unique_words) 
```

```python
# c)
# Iterate through each word and letter to find combinations 
vow_lib ={}
for w in list(count.keys()):
    vow_num = 0
    lett_seq = ''
    for l in range(len(w)):
        if w[l] in "aeiouy":
            vow_num += 1
        elif w[l] not in "aeiouy" and vow_num != 0:
            for n in range(vow_num):
                lett_seq += w[l-vow_num]
                vow_num -= 1
            counter(lett_seq,vow_lib)
            lett_seq = ''
print(len(vow_lib))
print(vow_lib)
```

#### 5. Make a Concordance

a. Create a concordance display for an arbitrary word. See the example below  

![](../../Data/figs/Sample-concordance-lines-of-actually.png)

```python
# a) 
def concordance(selected_word, words_surround, list_of_words):
    concordance_list = []
    concordance_sentence = ''
    for i in range(len(list_of_words)):
        if list_of_words[i] == selected_word and i - words_surround > 0 and i + words_surround < len(list_of_words):
            before = [list_of_words[i-j] for j in range(words_surround+1, 1, -1)]
            after = [list_of_words[i+k] for k in range(1, words_surround+1)]
            concordance_sentence = " ".join(before) + " " + selected_word + " " + " ".join(after)
            concordance_list.append(concordance_sentence)
            concordance_sentence= ''
    [print(n) for n in enumerate(concordance_list)]
                                   
concordance('apartment', 6, text.split())    
```
